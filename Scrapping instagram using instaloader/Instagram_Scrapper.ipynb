# Scrapping Instagram accounts

## Import  <span style="color:darkgoldenrod">Instaloader</span>:

The <span style="color:darkgoldenrod">Instaloader module</span> is a Python package having great functionalities to <span style="color:darkviolet">scrap instagram</span>, it‚Äôs functions can be used as command-line utility. <br>
The Instaloader key is used to <span style="color:darkviolet">download</span>:

## Import the <span style="color:darkgoldenrod">module</span>

```python
from instaloader import Instaloader, Profile
```

## Create an <span style="color:darkgoldenrod">object</span> of Instaloader class

```python
user = Instaloader()
```

## Load a <span style="color:darkgoldenrod">profile</span> from an Instagram handle

```python
profile = Profile.from_username(user.context, "zuck")
```

## Extract some <span style="color:darkgoldenrod">valuable information</span> from an Instagram profile:

```python
print("Username: ", profile.username)
print("User ID: ", profile.userid)
print("Number of Posts: ", profile.mediacount)
print("Followers: ", profile.followers)
print("Followees: ", profile.followees)
print("Bio: ", profile.biography, profile.external_url)
```

```python
posts_sorted_by_likes = sorted(profile.get_posts(), key=lambda post: post.likes,reverse=True)
for post in posts_sorted_by_likes:
    user.download_post(post, "zuck")
```
